```python
#imports
import numpy as np
import matplotlib.pyplot as plt
#function pca
def pca(X): 
    
   # normalize the features
    X = (X - X.mean()) / X.std()

   # compute the covariance matrix
    X = np.matrix(X)
    cov = (X.T * X) / X.shape[0]
#    print('cov \n', cov)
#    print()
   # perform SVD
    U, S, V = np.linalg.svd(cov) # singular value decomposition (linalg=line algebre)
    return U, S, V

def project_data(X, U, k):
    U_reduced = U[:,:k]
    return np.dot(X, U_reduced)



def recover_data(Z, U, k):
    U_reduced = U[:,:k]
    return np.dot(Z, U_reduced.T)
```

```python
# Apply PCA on faces
from scipy.io import loadmat
faces = loadmat('C:\\machine learning\\data\\unsupervised\\ex7faces.mat')
X = faces['X']
print(X.shape)
plt.imshow(X)
```

```python
#show one face
face = np.reshape(X[41,:], (32, 32))
plt.imshow(face)
```

```python
U, S, V = pca(X)
Z = project_data(X, U, 1)
X_recovered = recover_data(Z, U, 1)
face = np.reshape(X_recovered[41,:], (32, 32))
plt.imshow(face)
 
```
